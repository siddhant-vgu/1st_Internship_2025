List

```python
Mylist = ["Apple","Banana","Orange","Hemu","Kubilal","Hemu"]
print(Mylist)
type(Mylist)

#Indexing...
print(Mylist[0])
print(Mylist[1])
print(Mylist[2])

print(Mylist[1:2])
print(Mylist[0:])
print(Mylist[:])
print(Mylist[:3])

# Updation (functions)...
Mylist.append("Mango")
print(Mylist)

Mylist.insert(4,"Siddhant")
print(Mylist)

Mylist.count("Hemu")

Mylist.sort()
print(Mylist)

Mylist.pop(2)

Mylist.remove("Hemu")
print(Mylist)
```

Set

```python
myset = {1,2,3,4,5,1,6}

type(myset)
print(myset)


#operations...
set1 = {12,1,54,76,13}
set2 = {13,2,3,45,21}

print(set1|set2) #union
print(set1&set2) #intersection
print(set1-set2)
print(set2-set1)


print(set1.union(set2))
print(set1.intersection(set2))

set1.clear()
print(set1)
```
