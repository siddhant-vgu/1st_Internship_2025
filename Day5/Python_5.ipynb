```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.title("Student hrs vs marks")
plt.xlable("Hrs")
plt.ylable("Marks")
plt.show()
```

```python
subject = ["M1","M2","chemistry","physics"]
marks = ["15","19","11",'16']

plt.bar(subject,marks)
plt.title("subject wise marks")
plt.xlable("Subject")
plt.ylable("Marks")
plt.show()
```

```python
subject = ["M1","M2","chemistry","physics"]
marks = ["15","19","11",'16']

plt.scatter(subject,marks)
plt.title("subject wise marks")
plt.xlable("Subject")
plt.ylable("Marks")
plt.show()
```
