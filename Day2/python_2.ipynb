```python
balance = 50000
pin = 1234

user_pin = int(input("Enter your pin: "))

if user_pin == pin:
  print(f"Your balance is: {balance}")
  amount = int(input("Enter the amount to withdraw: "))
  print(f"Withdraw {amount} successfully.")
  balance -= amount
  print(f"Your balance is: {balance}")
else:
  print("Invalid pin")
```
