```python
def print_name():
  name = input("Enter your name: ")
  print(f"Hello Mr. {name}")
def name_as_argument(a):
  print(f"Hello Mr. {a}")

print_name()
name_as_argument("Siddhant")
```

```python
def multi_argument(*arg):
  print(arg)

multi_argument(1,2,3,4,5,6,7,8,9)
```

```python
def multi_argument(*arg):
  print(sum(arg))

multi_argument(1,2,3,4,5)
```

```python
def Add(a,b,c):
  return a+b+c
Add(12,12,12)
```

```python
def student(**info):
  for key , value in info.items():
    print(f"{key} : {value}")

student(Name = "Tom", Contact = 7023446224 , Age = 19,Collage = "Vgu jaipur")
```

```python
num1 = float(input("Enter 1st number: "))
num2 = float(input("Enter 2nd number: "))
add = lambda num1,num2: num1 + num2

print(add(num1,num2))
```

```python
num1 = float(input("Enter 1st number: "))
check = lambda num1:"Even" if num1 % 2 == 0 else "Odd"

print(check(num1))
```
